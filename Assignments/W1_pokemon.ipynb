### 7. Which is the second fastest non-legendary 'Ghost' type pokemon from 4th generation ?

```python
#your code here
# 2nd max speed, non-legendary, Ghost type, gen 4
pokemon.head()
# pokemon['Type 1'].value_counts()['Ghost']
# pokemon['Type 2'].value_counts()['Ghost']
# so Type 1 & Type 2 have things in common

l = []
for i, j in pokemon.iterrows():
    if (j.Legendary==False) and (j.Generation==4):
        if (j['Type 1']=='Ghost') or (j['Type 2']=='Ghost'):
            l.append((j.Speed, i))
            
l.sort(reverse=True)
l
```

### 8. How many non-legendary pokemons have stronger defence but weaker attack than Charizard?

```python
#your code here
pokemon.head()

pokemon.index.value_counts()['Charizard']
# pokemon.loc(['Charizard'])
## Selecting row by index:
# iloc - selects rows based on an integer index
# loc - selects rows based on a labeled index

atk, dfn = pokemon.loc[['Charizard']].Attack, pokemon.loc[['Charizard']].Defense
# print(atk, dfn)

count=0
for i, j in pokemon.iterrows():
    if(j.Legendary==False):
#         print(j.Defense)
        if ((j.Defense > dfn).any() and (j.Attack < atk).any()):   # .any() kyu?
            count += 1

count
```

### How many generation-3, non-legendary pokemons have higher HP than the weakest generation-6, legendary pokemon?

```python
#your code here
l=[]
for i,j in pokemon.iterrows():
    if (j.Generation==6):
        l.append(j.HP)
        
l.sort(reverse=True)
weak = l[0]

count=0
for i,j in pokemon.iterrows():
    if (j.Generation==3) and (j.HP > weak):
        count += 1
        
count
```
